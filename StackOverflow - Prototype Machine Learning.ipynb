# Preliminary Machine Learning - Prototype

After a talk with one of our mentors, I was convinced that I needed a product done fast and then I could improve it with time. So, I focused on having something working fast. 

As out on the project idea, I could do 2 types of recommendation:

- User based. Recommend things users alike saw/commented/questioned
- Content based. Get the questions, comments, answered that would be closer to the clients interest (or tags)

As the first approach would be faster for a prototype and idea prove of concept (POC). This is the place I went. After reading about clustering I decided to run a few models on my 1k tags dataset and see how it performed. 

So far I had a data frame with the users and a tag, and the description of the tags. I need to create a sparse matrix to use on the machine learning algorithms found in scikit-learn

```python
# read the files downloaded from stack api
# loaded a file with my functions first
%loadpy data_preparation.py
tags = import_top_users_tags()
```

```python
# create the sparse matrix of the top 1k
tag_cut = 1000
tags_count = tags.groupby(['name'])['count'].sum()
top_tags = tags_count[tags_count.rank(ascending=False)<=tag_cut].index
top_tags_data = tags[tags.name.isin(top_tags)]
# widen the data
top_tags_wide = top_tags_data.pivot_table(['count'], index=['user_id'],
                                              columns=['name'],
                                              aggfunc=lambda x: np.sum(x))
top_tags_wide = top_tags_wide.fillna(0)
top_tags_wide.columns = ["{0}".format(l2) for l1, l2 in top_tags_wide.columns]
top_tags_wide.head()
```

On semi-supervised algorithms for clustering, the number of clusters are As every question has at most 5 tags, we could think that 5 tags is enough to try to define something on Stack Overflow. If the 5 most important tags define a user, I could possible do 200 groups that would have 5 really important features and the rest would be "white noise"

As I was thinking about something that would be quick to train as well, I started with the Mini Batch K Means, a algorithm constructed to group documents and should be really fast, to be used real-time

```python
from sklearn.cluster import MiniBatchKMeans
import seaborn as sns
# seaborn config
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8, 4)})
np.random.seed(9221999)
```

```python
%%time
# training a model
n_clusters = 1000
model = MiniBatchKMeans(n_clusters = n_clusters, reassignment_ratio = 1)
ret = model.fit(top_tags_wide)
```

```python
# the results are the labels and centers of the groups/clusters
mbk_means_labels = ret.labels_
mbk_means_cluster_centers = ret.cluster_centers_
mbk_means_labels
```

```python
# to check what kind of recommendations it gives me, let's put the group back into the dataset
group_users = top_tags_wide.copy()
group_users['label'] = mbk_means_labels
```

```python
# checking the amount of people in each group
group_users.reset_index(inplace=True)
group_users.groupby('label')['user_id'].nunique().sort(ascending=False, inplace =False).plot(kind='bar', title='Distribution per group')
```

There are some groups that really have the majority of people. Maybe it makes sense, as there are more tags (and more popularity) in some problems/languages.

```python
total_tags_count_user = top_tags_wide.sum(axis=1)
total_tags_count_user.name='total'
top_tags_wide_normal = top_tags_wide.mul(100, axis=0).div(total_tags_count_user,
                                                       axis=0)
```

```python
%%time
# training a model
model_normal = MiniBatchKMeans(n_clusters = n_clusters, reassignment_ratio = 1)
ret = model_normal.fit(top_tags_wide_normal)
```

```python
# the results are the labels and centers of the groups/clusters
mbk_means_labels = ret.labels_
mbk_means_cluster_centers = ret.cluster_centers_
group_users = top_tags_wide.copy()
group_users['label'] = mbk_means_labels
group_users.reset_index(inplace=True)
group_users.groupby('label')['user_id'].nunique().sort(ascending=False, inplace =False).plot(kind='bar', title='Distribution per group')
```

Given names to the top 5 groups

```python
top = group_users.groupby('label')['user_id'].nunique().sort(ascending=False, inplace =False)[0:20]
```

```python
# print the top 10 groups and the most significant tags in order
for i in top.index:
    tmp_group = group_users[group_users['label']==int(i)] 
    total_person = tmp_group['user_id'].nunique()
    tmp_group.drop(['user_id', 'label'], axis=1, inplace=True)
    print 'Group {} - {}: {}'.format(i, total_person, ','.join(list(tmp_group.sum(axis=0).sort(ascending=False, inplace =False)[0:10].index)))
```

### Ploting the groups

In a high dimensional world (1k) is not easy to visualize grouping. Applying dimensional reduction techniques, we can simulate a grouping that would happen if the users were in a 2d space

```python
from sklearn.decomposition import TruncatedSVD
###############################################################################
# Visualize the results on PCA-reduced data
data = top_tags_wide_normal
reduced_data = TruncatedSVD(n_components=2).fit_transform(data)
kmeans = model_normal
kmeans.fit(reduced_data)
```

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .2     # point in the mesh [x_min, m_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() + 1, reduced_data[:, 0].max() - 1
y_min, y_max = reduced_data[:, 1].min() + 1, reduced_data[:, 1].max() - 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(20,10))
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=1,
            color='w', zorder=10)
plt.title('Tags clustering (TruncatedSVD-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
